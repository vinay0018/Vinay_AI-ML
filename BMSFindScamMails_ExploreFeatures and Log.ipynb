NEXT STEPS:

Remove correlated features


```python
# Create correlation matrix
cor_matrix = Data_df.corr().abs()
# print(cor_matrix)
```

```python
cor_matrix.shape
```

```python
type(cor_matrix)
```

```python
# print(cor_matrix.Shortining_Service[cor_matrix.Shortining_Service>0.6])
```

```python
# Select upper triangle of correlation matrix

upper = cor_matrix.where(np.triu(np.ones(cor_matrix.shape), k=1).astype(bool))
```

```python
upper
```

```python
upper['port']
```

```python
#Finding feature pairs having Correlation > 0.8 or <-0.8  and then dropping the High Correlated Features
```

```python
to_drop = [column for column in Data_df.columns if (any(upper[column] > 0.8) or any(upper[column] < -0.8)) ]
```

```python
to_drop
```

```python
Data_df_New=Data_df.drop(to_drop,axis=1)
Data_df_New
```

```python
# Splitting X and y into training and testing sets



from sklearn.model_selection import train_test_split




X_train, X_test, y_train, y_test = train_test_split(Data_df_New, Target, random_state=1, test_size=0.25)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# import the ML algorithm
from sklearn.linear_model import LogisticRegression
# from sklearn.linear_model import LogisticRegressionCV
```

```python
# Using K fold


from sklearn.model_selection import KFold

kf = KFold(n_splits=7, shuffle=True, random_state=100)
```

```python
# Checking the folds created

for train_idx, test_idx in kf.split(X_train):
    print(train_idx)
    print(test_idx)
    print('\n')
```

```python
model= LogisticRegression()
```

```python
from sklearn.model_selection import cross_val_score

results =  cross_val_score(estimator=model, 
                              X=X_train, 
                              y=np.ravel(y_train), 
                              cv=kf, 
                              scoring='accuracy')

print('K-Fold accuracy scores : \n', results)
print('Mean score : \n', results.mean())
```

```python
model.fit(X_test,np.ravel(y_test))
y_pred=model.predict(X_test)
```

```python
y_pred.shape, y_pred
```

```python
# Evaluate the model


# Testing Accuracy
model.score(X_test,y_test)
```

```python

# Training Accuracy
model.score(X_train,y_train)
```

```python
# CONFUSION MATRIX 

from sklearn.metrics import confusion_matrix

confusion_matrix(y_pred,y_test)
```

```python
#ROC Curve
from sklearn import metrics

y_pred_proba= model.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
```

```python
#Plot ROC curve
plt.plot(fpr,tpr)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

To get a better sense of what is going on inside the logistic regression model, visualize how our model uses the different features and which features have greater effect.

```python
coeff = list(model.coef_[0])
labels = list(X_train.columns)
features = pd.DataFrame()
features['Features'] = labels
features['importance'] = coeff
features.sort_values(by=['importance'], ascending=True, inplace=True)
features['positive'] = features['importance'] > 0  
features.set_index('Features', inplace=True)
features.importance.plot(kind='barh', figsize=(11, 6),color = features.positive.map({True: 'blue', False: 'red'}))
plt.xlabel('Importance')
```
